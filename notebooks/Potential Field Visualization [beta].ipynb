```python
def get_potential_scalar(x, y, k):
    Z_t = complex(x, y)
    P = 0
    for Z_s in facilities:
        Z = Z_s - Z_t
        radius = abs(Z)
        if radius >= 1:
            P += 1 / (radius**k)
    return P
```

```python
def get_potential_vector(x, y, k):
    Z_t = complex(x, y)
    P = complex(0, 0)
    for facility in facilities:
        Z_s = complex(facility[0], facility[1])
        Z = Z_s - Z_t
        radius = abs(Z)
        if radius >= 1:
            P += Z / (radius**k)
    return P
```

```python
def plot_potential_streamplot(X, Y, k):
    U = np.zeros((n, n))
    V = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            P = get_potential_vector(X[i][j], Y[i][j], k)
            U[i][j], V[i][j] = P.real, P.imag
    plt.streamplot(X, Y, U, V)
    plt.plot()
```
